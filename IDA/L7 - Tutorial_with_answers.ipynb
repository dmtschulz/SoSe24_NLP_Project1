### Task5: Handle missing values in all columns of the DataFrame

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df["deck"] = df["deck"].fillna(df["deck"].mode()[0])
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

estimator = lambda x: sum(x) / len(x) * 100

sns.barplot(x='pclass', y="survived", data=df, ax=ax1, estimator=estimator)
ax1.set_title('Survival Rate by Passenger Class')

sns.barplot(x='sex', y="survived", data=df, ax=ax2, estimator=estimator)
ax2.set_title('Survival Rate by Gender')

sns.barplot(x='embarked', y="survived", data=df, ax=ax3, estimator=estimator)
ax3.set_title('Survival Rate by Embarked')

sns.barplot(x='alone', y="survived", data=df, ax=ax4, estimator=estimator)
ax4.set_title('Survival Rate by Alone')

plt.show()
```
